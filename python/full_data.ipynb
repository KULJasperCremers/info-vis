```python
combined_data = {}

for country,elections in election_data.items():
    combined_data[country] = {
        "election_data": elections,
        "leaning_data": {
            "far-left": [],
            "left": [],
            "center-left": [],
            "center": [],
            "center-right": [],
            "right": [],
            "far-right": []
        }
    }
    
    for election in elections:
            for party in elections[election]:
                # print(f"Party {party}")
                for leaning, parties in leaning_data.items():
                        if party in parties:
                            # print(f"Found party: {party} in {leaning}")
                            if party not in combined_data[country]['leaning_data'][leaning]:
                                combined_data[country]['leaning_data'][leaning].append(party)
                                # print(f"Added party: {party} to {leaning}")
```

```python
# Get all unique parties from the election data
election_parties = set()
for country, elections in election_data.items():
    for election in elections:
        election_parties.update(elections[election])

# Get all unique parties from the combined data
combined_parties = set()
for country, data in combined_data.items():
    for parties in data['leaning_data'].values():
        combined_parties.update(parties)

# Compare the two sets
if election_parties == combined_parties:
    print("The sets of parties are identical.")
else:
    print("The sets of parties are not identical.")
    print("Parties in election data but not in combined data:", election_parties.difference(combined_parties))
    print("Parties in combined data but not in election data:", combined_parties.difference(election_parties))
```
