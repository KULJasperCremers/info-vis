```python
def check_percentage(data):
    for country in data:
        for year in data[country]:
            total = np.round(sum(data[country][year].values()))
            if total != 100:
                print(f"Total is not 100% for {country} in {year}: {total}")
```

```python
def parties(data, country):
  p = set()
  for year in data[country]:
    for party in data[country][year]:
      p.add(party)
  return p
```
